```python
import pandas as pd
from sklearn.preprocessing import StandardScaler

btc_data = pd.read_csv('btc.csv')
eth_data = pd.read_csv('eth.csv')
btc_data
```

```python
import matplotlib.pyplot as plt

# Create subplots
fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(10, 8))

# Plot btc_data
ax1.plot(btc_data['Price (GBP)'])
ax1.set_title('Bitcoin Price (GBP)')
ax1.set_xlabel('Index')
ax1.set_ylabel('Price (GBP)')

# Plot eth_data
ax2.plot(eth_data['Price (GBP)'], color="orange")
ax2.set_title('Ethereum Price (GBP)')
ax2.set_xlabel('Index')
ax2.set_ylabel('Price (GBP)')

# Adjust layout and display the plots
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt

# Create subplots
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 6))

# Plot btc_data histogram
ax1.hist(btc_data['Price (GBP)'], bins=5)
ax1.set_title('Bitcoin Price (GBP) - Histogram')
ax1.set_xlabel('Price (GBP)')
ax1.set_ylabel('Frequency')

# Plot eth_data histogram
ax2.hist(eth_data['Price (GBP)'], bins=5, color="orange")
ax2.set_title('Ethereum Price (GBP) - Histogram')
ax2.set_xlabel('Price (GBP)')
ax2.set_ylabel('Frequency')

# Adjust layout and display the plots
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt

# Count the occurrences of each class in btc_data and eth_data
btc_class_counts = btc_data['Class'].value_counts()
eth_class_counts = eth_data['Class'].value_counts()

# Get the unique classes
classes = sorted(set(btc_data['Class']) | set(eth_data['Class']))

# Create subplots
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 6))

# Plot bar chart for btc_data
ax1.bar(classes, [btc_class_counts.get(cls, 0) for cls in classes])
ax1.set_title('Bitcoin Data')
ax1.set_xlabel('Class')
ax1.set_ylabel('Counts')

# Plot bar chart for eth_data
ax2.bar(classes, [eth_class_counts.get(cls, 0) for cls in classes], color="orange")
ax2.set_title('Ethereum Data')
ax2.set_xlabel('Class')
ax2.set_ylabel('Counts')

# Adjust layout and display the plots
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt

# Create subplots
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 6))

# Plot horizontal box plot for btc_data
ax1.boxplot(btc_data['Price (GBP)'], vert=False)
ax1.set_title('Bitcoin Data')
ax1.set_xlabel('Price (GBP)')

# Plot horizontal box plot for eth_data
ax2.boxplot(eth_data['Price (GBP)'], vert=False)
ax2.set_title('Ethereum Data')
ax2.set_xlabel('Price (GBP)')

# Adjust layout and display the plots
plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Select columns with indexes 2, 3, 4, 5, and 6 from btc_data
btc_selected_columns = btc_data.iloc[:, [2, 3, 4, 5, 6]]

# Select columns with indexes 2, 3, 4, 5, and 6 from eth_data
eth_selected_columns = eth_data.iloc[:, [2, 3, 4, 5, 6]]

# Create the heatmap data for btc_data
btc_heatmap = btc_selected_columns.corr()

# Create the heatmap data for eth_data
eth_heatmap = eth_selected_columns.corr()

# Create subplots
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 6))

# Plot heatmap for btc_data
heatmap1 = ax1.pcolormesh(btc_heatmap, cmap='coolwarm')
ax1.set_title('Bitcoin Data')
ax1.set_xticks(np.arange(len(btc_heatmap.columns))+0.5)
ax1.set_yticks(np.arange(len(btc_heatmap.columns))+0.5)
ax1.set_xticklabels(btc_heatmap.columns, rotation=45)
ax1.set_yticklabels(btc_heatmap.columns)
plt.colorbar(heatmap1, ax=ax1, fraction=0.046, pad=0.04)

# Plot heatmap for eth_data
heatmap2 = ax2.pcolormesh(eth_heatmap, cmap='coolwarm')
ax2.set_title('Ethereum Data')
ax2.set_xticks(np.arange(len(eth_heatmap.columns))+0.5)
ax2.set_yticks(np.arange(len(eth_heatmap.columns))+0.5)
ax2.set_xticklabels(eth_heatmap.columns, rotation=45)
ax2.set_yticklabels(eth_heatmap.columns)
plt.colorbar(heatmap2, ax=ax2, fraction=0.046, pad=0.04)

# Adjust layout and display the plots
plt.tight_layout()
plt.show()
```

```python
from scipy.stats import skew, kurtosis

# Calculate skewness and kurtosis for btc_data
btc_price_skewness = skew(btc_data["Price (GBP)"])
btc_price_kurtosis = kurtosis(btc_data["Price (GBP)"])

# Calculate skewness and kurtosis for eth_data
eth_price_skewness = skew(eth_data["Price (GBP)"])
eth_price_kurtosis = kurtosis(eth_data["Price (GBP)"])

print("Bitcoin Data:")
print("Skewness:", btc_price_skewness)
print("Kurtosis:", btc_price_kurtosis)
print()
print("Ethereum Data:")
print("Skewness:", eth_price_skewness)
print("Kurtosis:", eth_price_kurtosis)
```
