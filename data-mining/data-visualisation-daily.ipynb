```python
import pandas as pd
from sklearn.preprocessing import StandardScaler

btc_data_daily = pd.read_csv('filtered_data/btc_daily_filtered.csv')
eth_data_daily = pd.read_csv('filtered_data/eth_daily_filtered.csv')
btc_data_hr = pd.read_csv('filtered_data/btc_hr_filtered.csv')
eth_data_hr = pd.read_csv('filtered_data/eth_hr_filtered.csv')
eth_data_daily.describe()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Create subplots
fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(10, 8))

# Plot btc_data
ax1.plot(btc_data_daily['Price (GBP)'])
ax1.set_title('Daily Bitcoin Price (GBP)')
ax1.set_xlabel('Index')
ax1.set_ylabel('Price (GBP)')

# Plot eth_data
ax2.plot(eth_data_daily['Price (GBP)'], color="orange")
ax2.set_title('Daily Ethereum Price (GBP)')
ax2.set_xlabel('Index')
ax2.set_ylabel('Price (GBP)')

# Adjust layout and display the plots
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt

# Create subplots
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 6))

# Plot btc_data histogram
ax1.hist(btc_data_daily['Price (GBP)'], bins=5)
ax1.set_title('Bitcoin Price (GBP) - Histogram')
ax1.set_xlabel('Price (GBP)')
ax1.set_ylabel('Frequency')

# Plot eth_data histogram
ax2.hist(eth_data_daily['Price (GBP)'], bins=5, color="orange")
ax2.set_title('Ethereum Price (GBP) - Histogram')
ax2.set_xlabel('Price (GBP)')
ax2.set_ylabel('Frequency')

# Adjust layout and display the plots
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt

# Count the occurrences of each class in btc_data and eth_data
btc_class_counts = btc_data_daily['Class'].value_counts()
eth_class_counts = eth_data_daily['Class'].value_counts()

# Get the unique classes
classes = sorted(set(btc_data_daily['Class']) | set(eth_data_daily['Class']))

# Create subplots
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 6))

# Plot bar chart for btc_data
ax1.bar(classes, [btc_class_counts.get(cls, 0) for cls in classes])
ax1.set_title('Bitcoin Data')
ax1.set_xlabel('Class')
ax1.set_ylabel('Counts')

# Plot bar chart for eth_data
ax2.bar(classes, [eth_class_counts.get(cls, 0) for cls in classes], color="orange")
ax2.set_title('Ethereum Data')
ax2.set_xlabel('Class')
ax2.set_ylabel('Counts')

# Adjust layout and display the plots
plt.tight_layout()
plt.show()
```

```python
from scipy.stats import skew, kurtosis

# Calculate skewness and kurtosis for btc_data
btc_price_skewness = skew(btc_data_daily["Price (GBP)"])
btc_price_kurtosis = kurtosis(btc_data_daily["Price (GBP)"])

# Calculate skewness and kurtosis for eth_data
eth_price_skewness = skew(eth_data_daily["Price (GBP)"])
eth_price_kurtosis = kurtosis(eth_data_daily["Price (GBP)"])

print("Bitcoin Data:")
print("Skewness:", btc_price_skewness)
print("Kurtosis:", btc_price_kurtosis)
print()
print("Ethereum Data:")
print("Skewness:", eth_price_skewness)
print("Kurtosis:", eth_price_kurtosis)
```

```python
import pandas as pd

# Load your Bitcoin and Ethereum data into btc_data and eth_data DataFrames
# Assuming you have 'Date' and 'Price' columns
btc_data_daily['Date'] = pd.to_datetime(btc_data_daily['Date'])
eth_data_daily['Date'] = pd.to_datetime(eth_data_daily['Date'])

window_sizes = [7,30]  # Define the rolling window sizes

for window in window_sizes:
    btc_data_daily[f'Rolling_SD_{window}'] = btc_data_daily["Price (GBP)"].rolling(window=window).std()
    eth_data_daily[f'Rolling_SD_{window}'] = eth_data_daily["Price (GBP)"].rolling(window=window).std()
    
# Create subplots
fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(10, 8))

# Plot rolling standard deviation for Bitcoin
for window in window_sizes:
    ax1.plot(btc_data_daily['Date'], btc_data_daily[f'Rolling_SD_{window}'], label=f'{window}-day Window')

ax1.set_title('Rolling Standard Deviation of Bitcoin Price')
ax1.set_xlabel('Date')
ax1.set_ylabel('Rolling Standard Deviation')
ax1.legend()

# Plot rolling standard deviation for Ethereum
for window in window_sizes:
    ax2.plot(eth_data_daily['Date'], eth_data_daily[f'Rolling_SD_{window}'], label=f'{window}-day Window')

ax2.set_title('Rolling Standard Deviation of Ethereum Price')
ax2.set_xlabel('Date')
ax2.set_ylabel('Rolling Standard Deviation')
ax2.legend()

# Adjust layout and display the plots
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load your dataset into a pandas DataFrame
df = pd.read_csv('filtered_data/btc_daily_filtered.csv')

# Assuming the target variable is in a column named 'Class'
target_column = 'Class'

# Count the number of instances in each class
class_counts = df[target_column].value_counts()

# Calculate class frequencies
class_frequencies = class_counts / len(df)

# Plot the class distribution
plt.figure(figsize=(8, 6))
class_frequencies.plot(kind='bar')
plt.xlabel('Class')
plt.ylabel('Frequency')
plt.title('Class Distribution')
plt.show()

# Compute class imbalance ratio (assuming binary classification)
imbalance_ratio = class_counts[0] / class_counts[1]

print('Class Counts:\n', class_counts)
print('Class Frequencies:\n', class_frequencies)
print('Imbalance Ratio:', imbalance_ratio)
```
